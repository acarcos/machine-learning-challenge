# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import StandardScaler
from numpy.random import seed
seed(1)
from tensorflow import random
random.set_seed(1)
```

```python
X_scaler = StandardScaler().fit(X_train)
print(X_scaler)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
# One-hot encoding
# Change to categorical
# THIS CANNOT BE USED FOR TEXT
#from tensorflow.keras.utils import to_categorical
#y_train_categorical = to_categorical(y_train)
#y_test_categorical = to_categorical(y_test)
```

```python
y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=X_train_scaled.shape[1]))
#model.add(Dense(units=50, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

#### You are passing a target array of shape (5243, 1) while using as loss `categorical_crossentropy`. `categorical_crossentropy` expects targets to be binary matrices (1s and 0s) of shape (samples, classes). If your targets are integer classes, you can convert them to the expected format via:

#### from keras.utils import to_categorical
#### y_binary = to_categorical(y_int)

#### Alternatively, you can use the loss function `sparse_categorical_crossentropy` instead, which does expect integer targets.

```python
# Fit the model to the training data
model.fit(X_train_scaled, y_train, epochs=200, shuffle='True', verbose=2)
```

```python
model2 = Sequential()
model2.add(Dense(units=4, activation='sigmoid', input_dim=X_train_scaled.shape[1]))
#model2.add(Dense(units=4, activation='relu'))
model2.add(Dense(units=3, activation='softmax'))
```
